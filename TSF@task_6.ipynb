# Trends and Seasonality Of TimeSeries Dataset

```python
total_cases = c_19_1[['total_cases']]
total_cases.rolling(12).mean().plot(figsize = (20 , 10), linewidth = 5, fontsize = 20)
plt.xlabel('Month' , fontsize = 20);
```

```python
new_cases = c_19_1[['new_cases']]
new_cases.rolling(12).mean().plot(figsize = (20 , 10), linewidth = 5, fontsize = 20)
plt.xlabel('Month' , fontsize = 20);
```

```python
sns.pairplot(c_19_1, size=2.5)
```

```python
with sns.axes_style('white'):
    sns.jointplot("new_cases", "total_cases", data=c_19_1, kind='reg')
```

```python
sns.distplot(c_19_1['total_cases'], kde=False);
plt.axvline(0, color="k", linestyle="--");
```

```python
sns.distplot(c_19_1['new_cases'], kde=False);
plt.axvline(0, color="r", linestyle="--");
```

```python
sns.distplot(c_19_1['total_tests'], kde=False);
plt.axvline(0, color="r", linestyle="--");
```

```python
sns.distplot(c_19_1['new_tests'], kde=False);
plt.axvline(0, color="r", linestyle="--");
```

```python
total_cases.diff().plot(figsize = (20 , 10), linewidth = 5, fontsize = 20)
plt.xlabel('Month', fontsize = 20);
```
